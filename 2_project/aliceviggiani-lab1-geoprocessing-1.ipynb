```python
import pandas as pd
import numpy as np
```

```python
# reading first dataset
# list of nyc public school with dual lenguage programs 

schools_list_path = "/Users/aliceviggiani/Dropbox/1 estudos/5 data viz/0 master/pratt/3-2025-fall/612-advanced_gis/lab1-geoprocessing/1-received/schools/bilingual-program-list.csv"
df_schools = pd.read_csv(schools_list_path)

df_schools
```

```python
# spliting the school addresses in more than one variable, so the ArcGIS Pro can find more precise localizations

df_schools[["Address", "Borough_check", "State", "Zipcode"]] = (
    df_schools["School Address"].str.split(",", expand=True).apply(
        lambda x: x.str.strip()
    )
)

df_schools
```

```python
# comparing results with Borough variable to check accuracy

df_schools["Borough_check"] = df_schools["Borough_check"].str.title()

df_check = df_schools[df_schools["Borough"] != df_schools["Borough_check"]]
df_check
```

```python
# manually normalizing and grouping languages to harmonize with another dataset later

df_schools['Language'] = df_schools['Language'].replace(
    {
    'Haitian Creole': 'French',
    'Bengali': 'Indo-European',
    'Italian': 'Indo-European', 
    'Yiddish': 'Indo-European',
    'Polish': 'Slavic',
    'Russian': 'Slavic',
    '(American) Sign Language': 'Other'
    }
)

df_schools['Language'].unique()
```

```python
# downloading spreadsheet to use in ArcGIS Pro

df_schools.to_csv("bilingual-program-list-5.csv", index=False)
```

```python
# back in Python:

# grouping the schools by language in each district
df_schools['schools_language'] = df_schools.groupby(['District', 'Language'])['School'].transform('count')
df_schools

# creating a new dataframe to organize data by school district
df_district = df_schools.groupby(['District', 'Language'])['School'].count().reset_index()
df_district.columns = ['district', 'language', 'school_sum']

df_district
```

```python
# pivot dataframe to create one observation by district

df_district_pivot = df_district.pivot(index='district', columns='language', values='school_sum')
df_district_pivot.columns = ['sch_' + col.lower().replace(' ', '_') for col in df_district_pivot.columns]

df_district_pivot = df_district_pivot.fillna(0).astype(int)
df_district_pivot = df_district_pivot.rename(columns={'sch_indo-european': 'sch_indo_european'})


df_district_pivot
```

```python
# adding blank variables for future interaction with population dataset

df_district_pivot['sch_asian'] = 0
df_district_pivot['sch_unspecified'] = 0

df_district_pivot
```

```python
# reading datasets about languages spoken at home in nyc created in ArcGIS Pro

languages_path = "/Users/aliceviggiani/Dropbox/1 estudos/5 data viz/0 master/pratt/3-2025-fall/612-advanced_gis/lab1-geoprocessing/3-produced/nyc_languages_tracts_3.csv"
df_languages = pd.read_csv(languages_path)

df_languages.columns.tolist()
```

```python
# filtering and renaming

df_languages = df_languages[[
    'Name',
    'State',
    'County',
    'school_district',
    'total_pop',
    'pop_spanish',
    'pop_french',
    'pop_indo_european',
    'pop_german',
    'pop_slavic',
    'pop_arabic',
    'pop_chinese',
    'pop_korean',
    'pop_vietnamese',
    'pop_tagalog',
    'pop_asian_others',
    'pop_unspecified',
    'pop_eng_only',
    'per_eng_only',
    'pop_bilingual_eng_another',
    'per_bilingual_eng_another',
    'pop_eng_limited_ability',
    'per_eng_limited_ability',
    'object_id',
    'TARGET_FID',
    'Geographic_Identifier___FIPS_Code',
    'Shape_Length',
    'Shape_Area'
]]

df_languages = df_languages.rename(
    columns={
    'school_district': 'School_district',
    'total_pop': 'Total',
    'pop_eng_only': 'Eng_only',
    'per_eng_only': 'Eng_only_per',
    'pop_indo_european': 'pop_indo_european_temp',
    'pop_bilingual_eng_another': 'Dual_language',
    'per_bilingual_eng_another': 'Dual_language_per',
    'pop_eng_limited_ability': 'Eng_limited_ability',
    'per_eng_limited_ability': 'Eng_limited_ability_per'        
    }
)

df_languages
```

```python
# uniformizing pop languages groups with schools languages

df_languages['pop_indo_european'] = df_languages[['pop_indo_european_temp', 'pop_german']].sum(axis=1)
df_languages['pop_asian'] = df_languages[['pop_vietnamese', 'pop_tagalog', 'pop_asian_others']].sum(axis=1)

# dropping columns
df_languages = df_languages.drop(['pop_indo_european_temp', 'pop_german', 'pop_vietnamese', 'pop_tagalog', 'pop_asian_others'], axis=1)

df_languages
```

```python
# joining datasets 
# school districts and languages spoken at home 

df_join = df_languages.merge(
    df_district_pivot, 
    left_on='School_district', 
    right_on='district', 
    how='left'
)

df_join
```

```python
df_join.columns.tolist()
```

```python
# creating a variable with the largest non-English native population for each tract

largest = [
    col for col in df_join.columns 
    if col.startswith('pop_')
]

# largest language
df_join['largest_language'] = df_join[largest].idxmax(axis=1)
df_join['largest_language'] = df_join['largest_language'].str.replace('pop_', '')

# largest value
df_join['largest_value'] = df_join[largest].max(axis=1)

# percentage
df_join['largest_per'] = ((df_join['largest_value'] / df_join['Total']) * 100).round(2)

df_join
```

```python
# creating variables with the proportion of population per school per language

languages = ['spanish', 'french', 'indo_european', 'slavic', 'arabic', 'chinese', 'korean', 'asian', 'unspecified']

for l in languages:
    df_join[f'prop_{l}'] = df_join[f'pop_{l}'] / df_join[f'sch_{l}'].replace(0, np.nan)
    df_join[f'prop_{l}'] = df_join[f'prop_{l}'].replace([np.inf, -np.inf], np.nan).fillna(0).round(2)

df_join
```

```python
# creating variable for the language with most schools per tract 

most = [
    col for col in df_join.columns 
    if col.startswith('sch_')
]

# more frequent school language
df_join['most_school'] = df_join[most].idxmax(axis=1)
df_join['most_school'] = df_join['most_school'].str.replace('sch_', '')

# value
df_join['most_value'] = df_join[most].max(axis=1)

df_join
```

```python
# creating variables for the language with lowest proportion population per school 

prop_l = [
    col for col in df_join.columns 
    if col.startswith('prop_')
]

# # more frequent school language
# df_join['prop_lower_language'] = df_join[prop_l].idxmin(axis=1)

# # value
# df_join['prop_lower'] = df_join[prop_l].min(axis=1)

# more frequent school language
df_join['prop_lower_language_nan'] = df_join[prop_l].replace(0, float('nan')).idxmin(axis=1)

# value
df_join['prop_lower_value_nan'] = df_join[prop_l].replace(0, float('nan')).min(axis=1)


# polishing 
df_join['prop_lower_language_nan'] = df_join['prop_lower_language_nan'].str.replace('prop_', '')

df_join
```

```python
# creating one variable for the proportion of population per school that attends the major language
# relating variables "largest-language" and "prop_"

df_join['prop_major_language_value'] = df_join.apply(
    lambda row: row['prop_' + row['largest_language']], 
    axis=1
)

# # eliminating the results 0 that represents actualy 0 schools
# # keeping the results 0 that represents a tract that has dual language program even with no population speaking the respective language
# # relating columns "prop_major_language_value" and "most_value"
# df_join.loc[df_join['most_value'].isna(), 'prop_major_language_value'] = 999999

df_join
```
